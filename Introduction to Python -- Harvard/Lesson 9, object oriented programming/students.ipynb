```python
students = [
    {"name": "Hermione", "house": "Gryffindor"},
    {"name": "Harry", "house": "Gryffindor"},
    {"name": "Ron", "house": "Gryffindor"},
    {"name": "Draco", "house": "Slytherin"},
    {"name": "Padma", "house": "Ravenclaw"}
]

houses = []

# I put all the houses inside the houses list
for student in students:
    if student["house"] not in houses:
        houses.append(student["house"])

# I print every houses with this method because if not this will return me:
# ['Gryffindor', 'Slytherin', 'Ravenclaw']
for house in sorted(houses):
    print(house)
```

```python
students = [
    {"name": "Hermione", "house": "Gryffindor"},
    {"name": "Harry", "house": "Gryffindor"},
    {"name": "Ron", "house": "Gryffindor"},
    {"name": "Draco", "house": "Slytherin"},
    {"name": "Padma", "house": "Ravenclaw"}
]

# The fucntion set() will retrun me some objects in python
# set() is useful for delete duplicates elements in a list
houses = set()

# I put all the houses inside the houses list
for student in students:
    houses.add(student["house"])

# I print every houses with this method because if not this will return me:
# {'Gryffindor', 'Slytherin', 'Ravenclaw'}
for house in sorted(houses):
    print(house)
```

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

union_set = set1 | set2
print(union_set)  # Output: {1, 2, 3, 4, 5}

intersection_set = set1 & set2
print(intersection_set)  # Output: {3}

difference_set = set1 - set2
print(difference_set)  # Output: {1, 2}

symmetric_difference_set = set1 ^ set2
print(symmetric_difference_set)  # Output: {1, 2, 4, 5}


a = {1, 2}
b = {1, 2, 3}

print(a.issubset(b))  # Output: True
print(b.issuperset(a))  # Output: True
print(a.isdisjoint({4, 5}))  # Output: True
```

```python
balance  = 0


def main():
    print("Balance:", balance)
    deposit(100)
    withdraw(50)
    print("Balance:", balance)


def deposit(n):
    global balance
    balance += n


def withdraw(n):
    global balance
    balance -= n


main()
```

```python
class Account:
    def __init__(self):
        self._balance = 0
    
    @property
    def balance(self):
        return self._balance
    
    def deposit(self, n):
        self._balance += n

    def withdraw(self, n):
        self._balance -= n


def main():
    account = Account()
    print("Balance:", account.balance)
    account.deposit(100)
    account.withdraw(50)
    print("Balance:", account.balance)

main()
```

```python
def meow(n: int) -> None:
    """
    Meow n times
    
    :param n: Number of times to meow
    :type n: int
    :raise TypeError: If n is not in int
    :return: A string of n meows, one per line
    :rtype: str
    """
    return "meow\n" * n


number: int = int(input("Number: "))
meows: str = meow(number)
print(meows, end="")
```
